```python
# OpenWeatherMap API Key
api_key = api_keys.api_key
city_weather_data = []
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
set_count = 1
record_count = 0
print("- - - - - Retrive data begin!!!!- - - - - - - - ")
for i,citi in enumerate(cities):
    record_count += 1
    if (i%50 == 0 and i>=50):
        record_count = 1
        set_count += 1
    print("Processing Record %s of Set %s | %s" % (record_count,set_count,citi))
    
    try:
        city_weather = requests.get(url+"&q="+citi).json()
        city_weather_data.append({
            "City":citi,
            "Country" : city_weather["sys"]["country"],
            "Date" : city_weather["dt"],
            "Latitude" : city_weather["coord"]["lat"],
            "Longitude" : city_weather["coord"]["lon"],
            "Humidity" : city_weather["main"]["humidity"],
            "Temperature" : city_weather["main"]["temp"],
            "Cloudiness" : city_weather["clouds"]["all"],
            "Wind_Speed" : city_weather["wind"]["speed"]
            
            
            
            
        })
    
    
    except:
        print("No such city")
        pass
        

print("- - - - - - - - - - - - - - -")
print("Data Retrieval Complete      ")
print("- - - - - - - - - - - - - - -")
```

```python
# Store the data into Dataframe
weather_data = pd.DataFrame(city_weather_data)
weather_data.head()
```

```python
# Output as csv file
weather_data.to_csv(output_data_file, index_label = "City_ID")
```

```python
# Check data consistent
weather_data.count()
```

## Latitude vs Temperature

```python
# Build scatter plot for latitude vs temperature
plt.scatter(weather_data["Latitude"], 
            weather_data["Temperature"],
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Cities")

# Incorporate the other graph properties
plt.title("City Latitude vs. Temperature (%s)" % time.strftime("%x"))
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("output_data/Fig1.png")

# Show plot
plt.show()
```

## Latitude vs. Humidity Plot

```python
# Build scatter plot for latitude vs temperature
plt.scatter(weather_data["Latitude"], 
            weather_data["Humidity"],color="coral",
            edgecolor = "black",linewidths=1, marker="o", 
            alpha=0.8, label="Cities")

# Incorporate the other graph properties
plt.title("City Latitude vs. Humidity (%s)" % time.strftime("%x"))
plt.ylabel("Humidity")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("output_data/Fig2.png")

# Show plot
plt.show()
```
